```python
# -*- coding: utf-8 -*-
"""
Created on Wed Dec 25 17:46:13 2024

@author: Group 5 - Mathematics/Programming for AI Coureswork
"""

import numpy as np
import time
from matplotlib import pyplot as plt
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import seaborn as sns

##########################################
# Activation Functions
##########################################
# These are basic non-linear functions applied to the network's linear transformations.
# Also included are their derivatives, used during backpropagation.

def sigmoid(Z):
    return 1/(1+np.exp(-Z))

def deriv_sigmoid(Z):
    s = sigmoid(Z)
    return s*(1-s)

def ReLU(Z):
    return np.maximum(0, Z)

def deriv_ReLU(Z):
    # Gradient is 1 for positive values and 0 for negative
    return Z > 0

def leaky_relu(Z, alpha=0.01):
    # Leaky ReLU allows a small gradient for negative values, mitigating dead neurons
    return np.where(Z > 0, Z, alpha*Z)

def deriv_leaky_relu(Z, alpha=0.01):
    dZ = np.ones_like(Z)
    dZ[Z<0] = alpha
    return dZ

def softmax(Z):
    # Softmax converts logits into probabilities for multi-class classification
    exp_Z = np.exp(Z - np.max(Z, axis=0, keepdims=True))
    return exp_Z / np.sum(exp_Z, axis=0, keepdims=True)


##########################################
# Dropout - inverted dropout
##########################################
# Dropout randomly "drops" (sets to zero) some neurons during training to prevent overfitting.
# Inverted Dropout: During training, we scale the activations so that at test time,
# we don't need to scale them again.

def dropout_forward(A, keep_prob):
    # Create a mask of ones and zeros, where ones represent kept neurons
    mask = (np.random.rand(*A.shape) < keep_prob)
    # Scale activations by dividing by keep_prob to maintain expected output scale
    A_drop = A * mask / keep_prob
    return A_drop, mask

def dropout_backward(dA, mask, keep_prob):
    # Backprop through dropout just involves applying the same mask and scaling
    return dA * mask / keep_prob

##########################################
# Utilities
##########################################

def one_hot(Y, num_classes):
    # Convert class labels into one-hot encoded format
    one_hot_Y = np.zeros((num_classes, Y.size))
    one_hot_Y[Y, np.arange(Y.size)] = 1
    return one_hot_Y

def get_predictions(A):
    # Get predicted class as the index of the max probability for each example
    return np.argmax(A, axis=0)

def get_accuracy(A, Y):
    # Compare predictions with true labels and compute mean accuracy
    preds = get_predictions(A)
    return np.mean(preds == Y)

def compute_loss(A_out, Y, params, lambda_, reg_type, layer_count, num_classes):
    # Compute the cross-entropy loss plus any specified regularization (L1 or L2)
    m = Y.size
    one_hot_Y = one_hot(Y, num_classes)
    # Add small constant (1e-10) to avoid log(0)
    log_likelihood = -np.log(A_out[one_hot_Y==1]+1e-10)
    data_loss = np.sum(log_likelihood)/m

    # Regularization is used to prevent overfitting by penalizing large weights
    if lambda_ > 0:
        reg_loss = 0
        if reg_type == 'l2':
            # L2 regularization sums the squares of the weights
            for l in range(1, layer_count+1):
                W = params['W'+str(l)]
                reg_loss += 0.5 * lambda_ * np.sum(W*W)
        elif reg_type == 'l1':
            # L1 regularization sums the absolute values of the weights
            for l in range(1, layer_count+1):
                W = params['W'+str(l)]
                reg_loss += lambda_ * np.sum(np.abs(W))
        else:
            reg_loss = 0
        return data_loss + (reg_loss/m)
    else:
        return data_loss

##########################################
# Optimizers
##########################################
# Different algorithms to update parameters during training:
# - SGD: Vanilla Stochastic Gradient Descent
# - SGD with Momentum: Uses a velocity term to accelerate gradients
# - Adam: Combines ideas from RMSProp and Momentum, adaptive learning rates

class Optimizer:
    pass

class SGD(Optimizer):
    def __init__(self, alpha=0.01):
        self.alpha = alpha
    def update(self, params, grads):
        # Update each parameter directly by subtracting alpha * gradient
        L = len([k for k in params.keys() if k.startswith('W')])
        for l in range(1, L+1):
            params['W'+str(l)] -= self.alpha*grads['dW'+str(l)]
            params['b'+str(l)] -= self.alpha*grads['db'+str(l)]
        return params

class SGD_Momentum(Optimizer):
    def __init__(self, alpha=0.01, beta=0.9):
        # beta controls how much "momentum" we keep from previous updates
        self.alpha = alpha
        self.beta = beta
        self.cache = {}
    def update(self, params, grads):
        L = len([k for k in params.keys() if k.startswith('W')])
        for l in range(1, L+1):
            for var in ['W','b']:
                dvar = grads['d'+var+str(l)]
                v_key = 'v'+var+str(l)
                # Initialize the velocity if not present
                if v_key not in self.cache:
                    self.cache[v_key] = np.zeros_like(params[var+str(l)])
                # Update the velocity and then the parameter
                self.cache[v_key] = self.beta*self.cache[v_key] + (1-self.beta)*dvar
                params[var+str(l)] -= self.alpha*self.cache[v_key]
        return params

class Adam(Optimizer):
    def __init__(self, alpha=0.001, beta1=0.9, beta2=0.999, epsilon=1e-8):
        # Adam keeps track of both the first moment (m) and second moment (v) of gradients
        self.alpha = alpha
        self.beta1 = beta1
        self.beta2 = beta2
        self.epsilon = epsilon
        self.t = 0
        self.m = {}
        self.v = {}
    def update(self, params, grads):
        L = len([k for k in params.keys() if k.startswith('W')])
        self.t += 1
        for l in range(1, L+1):
            for var in ['W','b']:
                dvar = grads['d'+var+str(l)]
                key_m = 'm'+var+str(l)
                key_v = 'v'+var+str(l)

                # Initialize first and second moment variables if not present
                if key_m not in self.m:
                    self.m[key_m] = np.zeros_like(params[var+str(l)])
                    self.v[key_v] = np.zeros_like(params[var+str(l)])

                # Update biased first moment estimate
                self.m[key_m] = self.beta1*self.m[key_m] + (1-self.beta1)*dvar
                # Update biased second moment estimate
                self.v[key_v] = self.beta2*self.v[key_v] + (1-self.beta2)*(dvar**2)

                # Correct bias in moments
                m_corrected = self.m[key_m]/(1-self.beta1**self.t)
                v_corrected = self.v[key_v]/(1-self.beta2**self.t)

                # Update parameter
                params[var+str(l)] -= self.alpha * (m_corrected/(np.sqrt(v_corrected)+self.epsilon))
        return params

##########################################
# Fully Parametrizable NN Class
##########################################
# This class builds a fully parameterized MLP:
# - Dynamically set layer sizes, activations, optimizer, dropout, and regularization
# - Train (fit) the model on given data
# - Evaluate on validation and test sets
# - Includes early stopping

class FullyConnectedNN:
    def __init__(self,
                 layer_dims,
                 activations,
                 input_dim,
                 output_dim=10,
                 output_activation='softmax',
                 optimizer='adam',
                 alpha=0.001,
                 lambda_=0.0005,
                 reg_type='l2',
                 keep_prob=1.0,
                 lrelu_alpha=0.01,
                 epochs=50,
                 batch_size=128,
                 patience=10,
                 print_every=1):

        self.layer_dims = layer_dims
        self.activations = activations
        self.input_dim=input_dim,
        self.output_dim = output_dim
        self.output_activation = output_activation
        self.lambda_ = lambda_
        self.reg_type = reg_type
        self.keep_prob = keep_prob
        self.lrelu_alpha = lrelu_alpha
        self.epochs = epochs
        self.batch_size = batch_size
        self.patience = patience
        self.print_every = print_every
        self.alpha = alpha

        # input_dim is global (set above), representing the input features
        self.input_dim = input_dim
        self.params = {}
        # Initialize parameters (weights & biases) for each layer
        self.init_params()
        self.L = len(self.layer_dims)+1 # total layers = hidden layers + output layer

        # Select optimizer based on user choice
        if optimizer=='adam':
            self.optimizer = Adam(alpha=alpha)
        elif optimizer=='sgd':
            self.optimizer = SGD(alpha=alpha)
        elif optimizer=='sgd_momentum':
            self.optimizer = SGD_Momentum(alpha=alpha, beta=0.9)
        else:
            raise ValueError("Unknown optimizer")

    def init_params(self):
        # He initialization for weights: np.sqrt(2.0/dim)
        np.random.seed(42)
        dims = [self.input_dim] + self.layer_dims + [self.output_dim]
        for l in range(1, len(dims)):
            self.params['W'+str(l)] = np.random.randn(dims[l], dims[l-1]) * np.sqrt(2.0/dims[l-1])
            self.params['b'+str(l)] = np.zeros((dims[l],1))

    def forward_prop(self, X, training=True):
        # Forward propagation: compute outputs layer by layer
        # caching intermediate values (Z, A, mask) for backprop
        caches = []
        A = X

        # Forward through hidden layers
        for l in range(1, len(self.layer_dims)+1):
            W = self.params['W'+str(l)]
            b = self.params['b'+str(l)]
            Z = W.dot(A) + b

            # Select activation function dynamically
            if self.activations[l-1] == 'relu':
                A_act = ReLU(Z)
                d_act = deriv_ReLU(Z)
            elif self.activations[l-1] == 'lrelu':
                A_act = leaky_relu(Z, alpha=self.lrelu_alpha)
                d_act = deriv_leaky_relu(Z, alpha=self.lrelu_alpha)
            else:
                # Default to sigmoid
                A_act = sigmoid(Z)
                d_act = deriv_sigmoid(Z)

            mask = None
            # Apply dropout if training (not during evaluation)
            if training and self.keep_prob < 1.0:
                A_act, mask = dropout_forward(A_act, self.keep_prob)

            caches.append((A, Z, W, b, d_act, mask))
            A = A_act

        # Output layer (softmax)
        W = self.params['W'+str(self.L)]
        b = self.params['b'+str(self.L)]
        Z = W.dot(A) + b
        A_out = softmax(Z)
        caches.append((A, Z, W, b, None, None))

        return A_out, caches

    def back_prop(self, A_out, Y, caches):
        # Backpropagation: compute gradients for all layers
        # Uses chain rule to propagate errors backward
        grads = {}
        L = self.L
        m = Y.size
        one_hot_Y = one_hot(Y, self.output_dim)

        # Derivative of loss w.r.t output layer activations
        (A_prev, Z, W_last, b_last, _, _) = caches[-1]
        dZ_last = A_out - one_hot_Y

        # Compute gradient of W at output layer including regularization
        if self.reg_type=='l2':
            dW_last = (1/m)*dZ_last.dot(A_prev.T) + (self.lambda_/m)*W_last
        elif self.reg_type=='l1':
            dW_last = (1/m)*dZ_last.dot(A_prev.T) + (self.lambda_/m)*np.sign(W_last)
        else:
            dW_last = (1/m)*dZ_last.dot(A_prev.T)

        grads['dW'+str(L)] = dW_last
        grads['db'+str(L)] = (1/m)*np.sum(dZ_last, axis=1, keepdims=True)

        # Propagate the error to the previous layer
        dA_prev = W_last.T.dot(dZ_last)

        # Backprop through hidden layers
        for l in reversed(range(1,L)):
            (A_prev, Z, W, b, d_act, mask) = caches[l-1]
            # dZ = dA * derivative_of_activation
            dZ = dA_prev * d_act
            if mask is not None:
                dZ = dropout_backward(dZ, mask, self.keep_prob)

            # Compute gradients for W, b with regularization
            if self.reg_type=='l2':
                dW = (1/m)*dZ.dot(A_prev.T) + (self.lambda_/m)*W
            elif self.reg_type=='l1':
                dW = (1/m)*dZ.dot(A_prev.T) + (self.lambda_/m)*np.sign(W)
            else:
                dW = (1/m)*dZ.dot(A_prev.T)

            db = (1/m)*np.sum(dZ, axis=1, keepdims=True)

            grads['dW'+str(l)] = dW
            grads['db'+str(l)] = db

            # dA_prev for the next iteration of backprop
            dA_prev = W.T.dot(dZ)

        return grads

    def fit(self, X_train, y_train, X_val, y_val):
      m = X_train.shape[1]
      best_val_acc = 0
      patience_counter = 0
      train_accuracies = []
      val_accuracies = []
      train_losses = []
      val_losses = []

      start_time = time.time()

      for epoch in range(self.epochs):
          # Shuffle data each epoch
          idxs = np.arange(m)
          np.random.shuffle(idxs)
          X_train = X_train[:, idxs]
          y_train = y_train[idxs]

          # Process mini-batches
          for i in range(0, m, self.batch_size):
              X_batch = X_train[:, i:i+self.batch_size]
              y_batch = y_train[i:i+self.batch_size]

              # Forward pass
              A_out, caches = self.forward_prop(X_batch, training=True)
              # Backward pass
              grads = self.back_prop(A_out, y_batch, caches)
              # Parameter update
              self.params = self.optimizer.update(self.params, grads)

          # Evaluate on training and validation sets
          A_train, _ = self.forward_prop(X_train, training=False)
          train_acc = get_accuracy(A_train, y_train)
          train_loss = compute_loss(A_train, y_train, self.params, self.lambda_, self.reg_type, self.L, self.output_dim)

          A_val, _ = self.forward_prop(X_val, training=False)
          val_acc = get_accuracy(A_val, y_val)
          val_loss = compute_loss(A_val, y_val, self.params, self.lambda_, self.reg_type, self.L, self.output_dim)

          # Track metrics
          train_accuracies.append(train_acc)
          val_accuracies.append(val_acc)
          train_losses.append(train_loss)
          val_losses.append(val_loss)

          # Print progress
          if (epoch + 1) % self.print_every == 0:
              print(f"Epoch {epoch + 1}/{self.epochs} - Train Acc: {train_acc:.4f}, Val Acc: {val_acc:.4f}, Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}")

          # Early stopping
          if val_acc > best_val_acc:
              best_val_acc = val_acc
              best_params = {k: v.copy() for k, v in self.params.items()}
              patience_counter = 0
          else:
              patience_counter += 1

          if patience_counter >= self.patience:
              print("Early stopping triggered!")
              self.params = best_params
              break

      end_time = time.time()
      train_time = end_time - start_time
      print(f"Training completed in {train_time:.2f} seconds. Best Val Acc: {best_val_acc:.4f}")

      # Plot training/validation curves
      self.plot_results(train_accuracies, val_accuracies, train_losses, val_losses)
      return best_val_acc, train_time



    def plot_results(self, train_accuracies, val_accuracies, train_losses, val_losses):
        # Plot accuracy and loss curves to visualize training progress
        plt.figure(figsize=(10,4))
        plt.subplot(1,2,1)
        plt.plot(train_accuracies, label='Train Acc')
        plt.plot(val_accuracies, label='Val Acc')
        plt.xlabel('Epochs')
        plt.ylabel('Accuracy')
        plt.title('Train vs Val Accuracy')
        plt.legend()

        plt.subplot(1,2,2)
        plt.plot(train_losses, label='Train Loss')
        plt.plot(val_losses, label='Val Loss')
        plt.xlabel('Epochs')
        plt.ylabel('Loss')
        plt.title('Train vs Val Loss')
        plt.legend()

        plt.tight_layout()
        plt.show()

    def predict(self, X):
        # Predict class labels for given input data
        A_out, _ = self.forward_prop(X, training=False)
        return get_predictions(A_out)

    def evaluate(self, X, Y):
        # Evaluate accuracy on a given dataset
        A_out, _ = self.forward_prop(X, training=False)
        acc = get_accuracy(A_out, Y)
        return acc, A_out
```

```python
import numpy as np
from matplotlib import pyplot as plt
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
import seaborn as sns

# Load datasets
def load_datasets():
    # MNIST
    (X_train_mnist, y_train_mnist), (X_test_mnist, y_test_mnist) = tf.keras.datasets.mnist.load_data()
    X_train_mnist, X_test_mnist = X_train_mnist / 255.0, X_test_mnist / 255.0

    # Fashion-MNIST
    (X_train_fmnist, y_train_fmnist), (X_test_fmnist, y_test_fmnist) = tf.keras.datasets.fashion_mnist.load_data()
    X_train_fmnist, X_test_fmnist = X_train_fmnist / 255.0, X_test_fmnist / 255.0

    datasets = {
        "MNIST": (X_train_mnist, y_train_mnist, X_test_mnist, y_test_mnist, 28 * 28, 10),
        "Fashion-MNIST": (X_train_fmnist, y_train_fmnist, X_test_fmnist, y_test_fmnist, 28 * 28, 10),
    }

    return datasets

# Preprocess dataset
def preprocess_data(X_train, y_train, X_test, y_test, input_dim, val_ratio=0.1):
    # Flatten and transpose
    X_train = X_train.reshape(X_train.shape[0], -1).T
    X_test = X_test.reshape(X_test.shape[0], -1).T
    y_train = y_train.flatten()
    y_test = y_test.flatten()

    # Split training data into training and validation sets
    train_size = int(X_train.shape[1] * (1 - val_ratio))
    X_train, X_val = X_train[:, :train_size], X_train[:, train_size:]
    y_train, y_val = y_train[:train_size], y_train[train_size:]

    return X_train, y_train, X_val, y_val, X_test, y_test


def train_and_evaluate(datasets):
    results = {}
    for name, (X_train, y_train, X_test, y_test, input_dim, num_classes) in datasets.items():
        print(f"\nTraining on {name}...")

        # Preprocess data with validation split
        X_train, y_train, X_val, y_val, X_test, y_test = preprocess_data(X_train, y_train, X_test, y_test, input_dim)

        # Select model parameters based on dataset
        if name == "MNIST":
            model = FullyConnectedNN(
                layer_dims=[128, 64],
                activations=['relu', 'relu'],
                input_dim=input_dim,
                output_dim=num_classes,
                output_activation='softmax',
                optimizer='adam',
                alpha=0.001,
                lambda_=0.0001,
                reg_type='l2',
                keep_prob=0.9,
                epochs=20,
                batch_size=64,
                patience=5,
                print_every=1
            )
        elif name == "Fashion-MNIST":
            model = FullyConnectedNN(
                layer_dims=[256, 128],
                activations=['lrelu', 'relu'],
                input_dim=input_dim,
                output_dim=num_classes,
                output_activation='softmax',
                optimizer='adam',
                alpha=0.001,
                lambda_=0.0005,
                reg_type='l2',
                keep_prob=0.8,
                epochs=20,
                batch_size=64,
                patience=5,
                print_every=1
            )
        else:
            continue

        # Train model using validation data
        best_val_acc, train_time = model.fit(X_train, y_train, X_val, y_val)
        test_acc, _ = model.evaluate(X_test, y_test)

        results[name] = {
            "Best Validation Accuracy": best_val_acc,
            "Test Accuracy": test_acc,
            "Train Time (s)": train_time
        }

        print(f"{name} Test Accuracy: {test_acc:.4f}")

    return results

def plot_comparison(results):
    datasets = list(results.keys())
    val_accuracies = [results[ds]["Best Validation Accuracy"] for ds in datasets]
    test_accuracies = [results[ds]["Test Accuracy"] for ds in datasets]

    plt.figure(figsize=(12, 6))

    # Validation Accuracy
    plt.bar(np.arange(len(datasets)) - 0.2, val_accuracies, width=0.4, label='Validation Accuracy', color='skyblue')
    # Test Accuracy
    plt.bar(np.arange(len(datasets)) + 0.2, test_accuracies, width=0.4, label='Test Accuracy', color='orange')

    plt.xticks(np.arange(len(datasets)), datasets)
    plt.ylabel("Accuracy")
    plt.title("Comparison of Best Models Across Datasets")
    plt.legend()
    plt.show()


# Main
datasets = load_datasets()
results = train_and_evaluate(datasets)
plot_comparison(results)
```
